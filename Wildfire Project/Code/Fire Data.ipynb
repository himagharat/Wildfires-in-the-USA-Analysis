```python
from matplotlib import pyplot as plt
from scipy import stats
from scipy.stats import linregress
import numpy as np
import pandas as pd
```

```python
Acre_Fire = "C:\\UPenn\\Wildfire Project\\Wildfire Project\\total_wildland_fires_acres.csv"
```

```python
Acre_Fire_df = pd.read_csv(Acre_Fire)
```

```python
Acre_Fire_df
```

```python
Fire_plot =plt.scatter(Acre_Fire_df['Year'],Acre_Fire_df['Fires'], color='red')
plt.xlabel("years")
plt.ylabel("Fires")
```

```python
x_values = Acre_Fire_df['Year']
y_values = Acre_Fire_df['Fires']
```

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Years')
plt.ylabel('Fires')
print(f"The r-squared is: {rvalue**2}")
print (line_eq)
plt.show()
```
